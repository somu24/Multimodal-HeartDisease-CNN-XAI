```python
#importing required libraries

import pandas as pd
import numpy as np
import tensorflow as tf
import PIL
from PIL import Image
from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import Adam
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_curve, accuracy_score
from inspect import signature
import os
import joblib
from skimage.segmentation import quickshift
from skimage.color import rgb2gray
from lime import lime_image
from keras.preprocessing import image
from keras.models import load_model
from keras.optimizers import Adam
from keras.losses import CategoricalCrossentropy
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense,BatchNormalization
from sklearn.model_selection import train_test_split
import seaborn as sns
from lime import lime_tabular
import re
from sklearn.utils import resample
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, BatchNormalization, Input
from tensorflow.keras.models import Sequential
from sklearn.metrics import classification_report, confusion_matrix
from lime import lime_tabular

#xray code
train_val_data = pd.read_csv('/Users/shivamaniarunachalam/Desktop/ADT_project_datas/train_val_list.txt', header=None, names=['image_filename'])

def get_full_path(image_name):
    path_002 = os.path.join("/Users/shivamaniarunachalam/Desktop/ADT_project_datas/images/images/", image_name)
    if os.path.exists(path_002):
        return path_002
    else:
        return None


train_val_data['Image_full_path'] = train_val_data['image_filename'].apply(get_full_path)

train_val_data.dropna(subset=['Image_full_path'], inplace=True)

all_img_path_df = pd.DataFrame(train_val_data['Image_full_path'])


def load_labels_from_csv(csv_file):
    df = pd.read_csv(csv_file)
    dff = all_img_path_df

    labels_dict = dict(zip(dff['Image_full_path'], df['Finding Labels']))
    return labels_dict

csv_file = '/Users/shivamaniarunachalam/Desktop/ADT_project_datas/Data_Entry_2017_1.csv'
labels_dict = load_labels_from_csv(csv_file)


labels_dataframe= pd.DataFrame(list(labels_dict.items()), columns=['Image_full_path', 'Label'])

labels_dataframe = labels_dataframe[labels_dataframe['Label'] != 'No Finding']


labels_dataframe['Labels'] = labels_dataframe['Label'].str.split('|')


filtered_dataframe = labels_dataframe[labels_dataframe['Labels'].apply(len) == 1]


filtered_dataframe = filtered_dataframe.drop(columns=['Labels'])




train_df, val_df = train_test_split(filtered_dataframe, test_size=0.2, random_state=42)

train_df.shape


unique_labels = filtered_dataframe['Label'].unique()
num_unique_labels = len(unique_labels)

batch_size = 64


class CustomDataGen(tf.keras.utils.Sequence):

    def __init__(self, df, X_col, y_col,
                 batch_size,
                 input_size=(320, 320, 1),
                 shuffle=True):

        self.df = df.copy()
        self.X_col = X_col
        self.y_col = y_col
        self.batch_size = batch_size
        self.input_size = input_size
        self.shuffle = shuffle

        self.n = len(self.df)
        # Create label to index mapping
        self.label_to_index = {label: idx for idx, label in enumerate(self.df[y_col['Label']].unique())}
        self.n_classes = len(self.label_to_index)
        self.n_label = df[y_col['Label']].nunique()

        self.on_epoch_end()


    def on_epoch_end(self):
        if self.shuffle:
            self.df = self.df.sample(frac=1).reset_index(drop=True)

    def __get_input(self, img_path, target_size):



        image = tf.keras.preprocessing.image.load_img(img_path,color_mode='grayscale')
        image_arr = tf.keras.preprocessing.image.img_to_array(image)


        image_arr = tf.image.resize(image_arr,(target_size[0], target_size[1])).numpy()

        return image_arr/255.

    def __get_output(self, label):
        idx = self.label_to_index[label]
        return tf.keras.utils.to_categorical(idx, num_classes=self.n_classes)

    def __get_data(self, batches):
        # Generates data containing batch_size samples

        path_batch = batches[self.X_col['Image_full_path']]


        label_batch = batches[self.y_col['Label']]


        X_batch = np.asarray([self.__get_input(x, self.input_size[:2]) for x in path_batch])

        y_batch = np.asarray([self.__get_output(y) for y in label_batch])


        return X_batch, y_batch

    def __getitem__(self, index):

        batches = self.df.iloc[index * self.batch_size:(index + 1) * self.batch_size]
        X, y = self.__get_data(batches)
        return X, y

    def __len__(self):
        return int(np.ceil(self.n / self.batch_size))

input_size=(320, 320, 1)
target_size=(320,320,1)

num_classes = 14


train_generator = CustomDataGen(train_df,
                         X_col={'Image_full_path':'Image_full_path'},
                         y_col={'Label':'Label'},
                         batch_size=64, input_size=target_size,shuffle = True)

val_generator = CustomDataGen(val_df,
                       X_col={'Image_full_path':'Image_full_path'},
                       y_col={'Label':'Label'},
                       batch_size=64, input_size=target_size,shuffle = False)



input_shape = (320,320,1)

def functional_mlp():

    inp = Input(shape=input_shape)
    x = Conv2D(32, (3, 3), activation='relu')(inp)

    x = MaxPooling2D((2, 2))(x)


    x = Conv2D(64, (3, 3), activation='relu')(x)

    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(128, (3, 3), activation='relu')(x)

    x = MaxPooling2D((2, 2))(x)

    x = Flatten()(x)

    x = Dense(128, activation='relu')(x)

    neural_nw_out = Dense(num_classes, activation='softmax')(x)

    return Model(inputs=[inp],outputs=[neural_nw_out])

model = functional_mlp()

model.summary()

def categorical_cross_entropy_loss(Y_true,Y_hat):

    return tf.reduce_mean(tf.keras.losses.categorical_crossentropy(y_true=Y_true,y_pred=Y_hat))


optimizer = Adam(learning_rate=0.001)
epochs = 15
steps_per_epoch = len(train_df)//batch_size

for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")

    # Training phase
    step = 0
    total_loss_train = 0.0
    correct_predictions_train = 0
    total_samples_train = 0

    for X_train_mb, Y_train_mb in train_generator:
        with tf.GradientTape() as tape:
            Y_pred = model(X_train_mb, training=True)
            loss_value = categorical_cross_entropy_loss(Y_train_mb, Y_pred)

        gradients = tape.gradient(loss_value, model.trainable_weights)
        optimizer.apply_gradients(zip(gradients, model.trainable_weights))

        batch_size = X_train_mb.shape[0]
        total_loss_train += loss_value.numpy() * batch_size

        predicted_labels = np.argmax(Y_pred, axis=1)
        true_labels = np.argmax(Y_train_mb, axis=1)
        correct_predictions_train += np.sum(predicted_labels == true_labels)
        total_samples_train += batch_size

        print(f"Step {step}, Loss: {loss_value.numpy()}")

        step += 1
        if step >= steps_per_epoch:  # Ensure to break after completing one epoch
            break

    # Calculate training metrics
    epoch_loss_train = total_loss_train / total_samples_train
    epoch_accuracy_train = correct_predictions_train / total_samples_train

    print(f"Epoch {epoch+1}, Average Training Loss per sample: {epoch_loss_train}, Average Training Accuracy per sample: {epoch_accuracy_train}")

    # Validation phase
    total_loss_val = 0.0
    correct_predictions_val = 0
    total_samples_val = 0

    # for index in range(len(val_generator)):
    #     X_val_mb, Y_val_mb = val_generator[index]
    #     Y_pred_val = model(X_val_mb, training=False)
    #     loss_value_val = categorical_cross_entropy_loss(Y_val_mb, Y_pred_val)

    #     batch_size_val = X_val_mb.shape[0]
    #     total_loss_val += loss_value_val.numpy() * batch_size_val

    #     predicted_labels_val = np.argmax(Y_pred_val, axis=1)
    #     true_labels_val = np.argmax(Y_val_mb, axis=1)
    #     correct_predictions_val += np.sum(predicted_labels_val == true_labels_val)
        # total_samples_val += batch_size_val

    # Calculate validation metrics
    # epoch_loss_val = total_loss_val / total_samples_val
    # epoch_accuracy_val = correct_predictions_val / total_samples_val

    # print(f"Epoch {epoch+1}, Average Validation Loss per sample: {epoch_loss_val}, Average Validation Accuracy per sample: {epoch_accuracy_val}")




joblib.dump(model, 'model.joblib')


model = joblib.load('model.joblib')
model_save_path = '/Users/shivamaniarunachalam/Desktop/ADT_project_datas/Xray_model.h5'
model.save(model_save_path)

#Xai for Xray

 
# Load the saved model
model = load_model('/Users/shivamaniarunachalam/Desktop/ADT_project_datas/Xray_model.h5')
 
# Compile the model
loss_function = CategoricalCrossentropy()
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss=loss_function, metrics=['accuracy'])
 
# Define the image preprocessing functions
def preprocess_image(image_path, target_size=(320, 320)):
    img = image.load_img(image_path, target_size=target_size, color_mode='grayscale')
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    return img_array
 
def convert_to_grayscale(image_array):
    if image_array.ndim == 3:
        return rgb2gray(image_array)
    elif image_array.ndim == 4 and image_array.shape[-1] == 1:
        return np.squeeze(image_array, axis=-1)
    else:
        raise ValueError("Unsupported image format")
 
def quickshift_segmentation(image, **kwargs):
    if image.ndim == 3 and image.shape[-1] == 1:  # Convert grayscale to RGB
        image_rgb = np.repeat(image, 3, axis=-1)
    else:
        image_rgb = image
    return quickshift(image_rgb, **kwargs)
 
# Define the prediction function for LIME
def new_predict_fn(images):
    images = np.expand_dims(images, axis=-1)  # Ensure images are in correct shape
    return model.predict(images)
 
# Explain an image with LIME and plot heatmap
def explain_image_with_lime(model, image_path, input_size):
    img_array = preprocess_image(image_path, target_size=input_size)
    # Create a LIME image explainer
    explainer = lime_image.LimeImageExplainer()
    # Explain the image
    explanation = explainer.explain_instance(
        img_array[0],  # Use the first image in batch
        new_predict_fn,
        top_labels=1,
        hide_color=0,
        num_samples=1000,
        segmentation_fn=quickshift_segmentation
    )
    # Get the superpixels and mask
    temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=False)
    # Plot the original image
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img_array[0].squeeze(), cmap='gray')
    plt.title('Original Image')
 
    # Plot the heatmap
    plt.subplot(1, 2, 2)
    plt.imshow(temp, cmap='jet', alpha=0.6)  # Use 'jet' colormap for heatmap
    plt.title('LIME Explanation Heatmap')
    plt.colorbar()
 
    plt.show()
 
# Example usage
image_path = "/Users/shivamaniarunachalam/Desktop/ADT_project_datas/images/images/123.png"
explain_image_with_lime(model, image_path, (320, 320))
image_path = "/Users/shivamaniarunachalam/Desktop/ADT_project_datas/images/images/00001836_070.png"
explain_image_with_lime(model, image_path, (320, 320))

image_path = "/Users/shivamaniarunachalam/Desktop/ADT_project_datas/images/images/00001836_123.png"
explain_image_with_lime(model, image_path, (320, 320))


#ecg code



#### Loading the Dataset
 
# reading dataset
train_data = pd.read_csv("/Users/shivamaniarunachalam/Desktop/ADT_project_datas/mitbih_train.csv", header=None)
test_data = pd.read_csv("/Users/shivamaniarunachalam/Desktop/ADT_project_datas/mitbih_test.csv", header=None)
 
train_data[187] = train_data[187].astype('int')
 
### Experimenting with Dataset
 
print("list of classes :", train_data.iloc[:, 187].unique())
sns.catplot(x=187, kind='count', data=train_data)
 
from sklearn.utils import resample
 
df_1 = resample(train_data[train_data[187] == 1], n_samples=10000, replace=True, random_state=123)
df_2 = resample(train_data[train_data[187] == 2], n_samples=10000, replace=True, random_state=123)
df_3 = resample(train_data[train_data[187] == 3], n_samples=10000, replace=True, random_state=123)
df_4 = resample(train_data[train_data[187] == 4], n_samples=10000, replace=True, random_state=123)
 
# downsample the high number of counts in one class, select random samples 2000 samples from class 0 samples
df_0 = train_data[train_data[187] == 0].sample(n=10000, random_state=123)
 
train_df = pd.concat([df_0, df_1, df_2, df_3, df_4]).sample(frac=1, random_state=42).reset_index(drop=True)
 
sns.catplot(x=187, kind='count', data=train_df)
 
from tensorflow.keras.utils import to_categorical
 
y_train = to_categorical(train_df[187])
y_test = to_categorical(test_data[187])
 
X_train = train_df.iloc[:, 1:187].values
X_test = test_data.iloc[:, 1:187].values
 
# For conv1D dimensionality should be 186 1 where 186 is number of features and 1 = 1D Dimensionality of data
X_train = X_train.reshape(len(X_train), X_train.shape[1], 1)
X_test = X_test.reshape(len(X_test), X_test.shape[1], 1)
 
## Model
 
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, BatchNormalization, Input
from tensorflow.keras.models import Sequential
 
model = Sequential([
    Conv1D(filters=64, kernel_size=6, activation='relu', padding='same', input_shape=(186, 1)),
    BatchNormalization(),
    MaxPooling1D(pool_size=3, strides=2, padding="same"),
 
    Conv1D(64, kernel_size=3, activation='relu'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2, strides=2, padding="same"),
 
    Conv1D(64, kernel_size=3, activation='relu'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2, strides=2, padding="same"),
 
    Conv1D(128, kernel_size=3, activation='relu'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2, strides=2, padding="same"),
 
    Flatten(),
 
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(5, activation='softmax')
])
 
# Display the model summary
model.summary()
 
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
 
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
 
model.evaluate(X_test, y_test)
 
model.save("model.h5")
 
# evaluate ECG Test Data
 
pd.DataFrame(history.history)
pd.DataFrame(history.history)[['accuracy', 'val_accuracy']].plot()
pd.DataFrame(history.history)[['loss', 'val_loss']].plot()
 
from sklearn.metrics import classification_report, confusion_matrix
 
# distributional probability to integers
yhat = np.argmax(model.predict(X_test), axis=1)
 
confusion_matrix(np.argmax(y_test, axis=1), yhat)
 
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(np.argmax(y_test, axis=1), yhat), annot=True, fmt='0.0f', cmap='RdPu')
 
## Interpreting with Lime
 
# Select an instance to explain
z = 19142
instance = X_test[z].reshape(1, 186)
 
# Assuming model is already trained and X_test is defined
explainer = lime_tabular.LimeTabularExplainer(
    training_data=X_train.reshape(X_train.shape[0], -1),  # Flatten the input
    mode="classification",
    feature_names=[f"feature_{i}" for i in range(X_train.shape[1])],
    class_names=[str(i) for i in range(5)],
    verbose=True,
    random_state=42
)
 
# prediction function
predict_fn = lambda data: model.predict(data.reshape(-1, 186, 1))
 
explanation = explainer.explain_instance(instance[0], predict_fn, top_labels=5, num_features=10, num_samples=1000)
 
# Choose explainer for predicted disease
top_label = explanation.top_labels[0]
exp = explanation.as_list(label=top_label)
 
# Array with influential values(scaled)
influence = np.zeros(instance.shape[1])
 
# Function to extract feature index from the feature string
def extract_feature_index(feature_str):
    # Find all numbers in the string
    numbers = re.findall(r'\d+', feature_str)
    if numbers:
        return int(numbers[0])  # Return the first number found
    else:
        raise ValueError(f"Could not find a number in the feature string: {feature_str}")
 
# Assign influence values to the corresponding points
for feature, weight in exp:
    feature_num = extract_feature_index(feature)
    influence[feature_num] = weight
 
# Normalize influence values for colormap
if np.max(influence) - np.min(influence) != 0:
    normalized_influence = (influence - np.min(influence)) / (np.max(influence) - np.min(influence))
else:
    normalized_influence = np.zeros(influence.shape)
 
# Plot the ECG signal with a background colored based on influence
fig, ax = plt.subplots(figsize=(15, 5))  # Create a figure and axes
 
cmap = plt.get_cmap('jet')  # Use the 'jet' colormap for a similar effect
 
# Create horizontal gradient background
for i in range(len(instance[0])):
    if normalized_influence[i] > 0:  # Highlight areas with influence
        color = cmap(normalized_influence[i])
        ax.axvspan(i - 1, i + 1, color=color, alpha=0.2)  # Increase width and set higher transparency
 
# Plot the actual ECG signal in black for reference
ax.plot(range(len(instance[0])), instance[0], color='black')
 
ax.set_title('ECG Signal with Influential Data Points Highlighted')
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
 
# Create a ScalarMappable and add a colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=np.min(influence), vmax=np.max(influence)))
sm.set_array([])
 
fig.colorbar(sm, ax=ax, label='Influence')  # Specify the axes
 
plt.show()
 
# Print the actual label for the instance
print(y_test[z])

#heart attack code

# Loading the dataset
heart_data = pd.read_csv('/Users/shivamaniarunachalam/Desktop/ADT_project_datas/extended_heart_disease_dataset.csv')

# Renaming columns for better readability
heart_data.rename(columns={
    'cp': 'chest_pain_type',
    'trestbps': 'resting_BP',
    'chol': 'Cholesterol',
    'fbs': 'fasting_blood_sugar>120',
    'restecg': 'resting_ecg',
    'thalach': 'max_hr',
    'exang': 'exercise_induced_angina',
    'oldpeak': 'std_depression',
    'slope': 'std_slope',
    'ca': 'num_major_vessels',
    'thal': 'thalassemia'
}, inplace=True)

# Mapping categorical values to meaningful labels
heart_data['chest_pain_type'] = heart_data['chest_pain_type'].map({0: 'typical', 1: 'atypical', 2: 'non-anginal', 3: 'asymptomatic'})
heart_data['resting_ecg'] = heart_data['resting_ecg'].map({0: 'normal', 1: 'st-t_wave_abnormality', 2: 'left_ventricular_hypertrophy'})
heart_data['std_slope'] = heart_data['std_slope'].map({0: 'upsloping', 1: 'flat', 2: 'downsloping'})
heart_data['thalassemia'] = heart_data['thalassemia'].map({3: 'reversable_defect', 2: 'fixed_defect', 1: 'normal'})

# Separating features (X) and target variable (Y)
X = heart_data.drop(['target'], axis=1)
Y = heart_data['target']

# Convert categorical variables to dummy/indicator variables
X = pd.get_dummies(X)

# Splitting the dataset into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Scaling the features for better performance of the model
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Hyperparameter tuning using Grid Search with Cross-Validation
param_grid = {
    'C': [0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs'],
    'max_iter': [100, 200, 300]
}

grid_search = GridSearchCV(LogisticRegression(random_state=0), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_scaled, Y_train)

# Output the best parameters and best accuracy score from Grid Search
print("Best parameters found: ", grid_search.best_params_)
print("Best accuracy found: ", grid_search.best_score_)

# Using the best classifier from the Grid Search
best_classifier = grid_search.best_estimator_

# Evaluate model performance using Cross-Validation
cv_scores = cross_val_score(best_classifier, X_train_scaled, Y_train, cv=10, scoring='accuracy')
print("Cross-validation scores: ", cv_scores)
print("Mean cross-validation score: ", np.mean(cv_scores))

# Train the model on the full training dataset
best_classifier.fit(X_train_scaled, Y_train)

# Making predictions on the test set
Y_pred = best_classifier.predict(X_test_scaled)

# Calculating accuracy on the test set
print("Accuracy on test set:", accuracy_score(Y_test, Y_pred))

# Plotting the Precision-Recall curve
precision, recall, _ = precision_recall_curve(Y_test, Y_pred)

step_kwargs = ({'step': 'post'} if 'step' in signature(plt.fill_between).parameters else {})
plt.step(recall, precision, color='b', alpha=0.2, where='post')
plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve')
plt.show()


# Model Interpretation using LIME
explainer = lime_tabular.LimeTabularExplainer(
    training_data=X_train_scaled,
    feature_names=X.columns,
    class_names=['No Heart Attack', 'Heart Attack'],
    mode='classification'
)

# Example new individual data for prediction
new_data = {
    'age': 55,
    'sex': 1,
    'chest_pain_type': 'non-anginal',
    'resting_BP': 140,
    'Cholesterol': 230,
    'fasting_blood_sugar>120': 0,
    'resting_ecg': 'normal',
    'max_hr': 150,
    'exercise_induced_angina': 0,
    'std_depression': 2.3,
    'std_slope': 'upsloping',
    'num_major_vessels': 0,
    'thalassemia': 'fixed_defect'
}

# Convert the new individual data to DataFrame
new_data_df = pd.DataFrame([new_data])

# Convert categorical features to dummy variables
new_data_df = pd.get_dummies(new_data_df)

# Ensure new data has the same columns as training data
missing_cols = set(X.columns) - set(new_data_df.columns)
for col in missing_cols:
    new_data_df[col] = 0

# Reorder columns to match the training set
new_data_df = new_data_df[X.columns]

# Scale the new data using the same scaler as the training data
new_data_scaled = scaler.transform(new_data_df)

# Predict the outcome for the new individual
new_prediction = best_classifier.predict(new_data_scaled)

# Output the prediction result
prediction_text = "will have a heart attack" if new_prediction[0] == 1 else "will not have a heart attack"
print(f"The new individual {prediction_text}.")

# Explaining the prediction for the new individual with LIME
exp = explainer.explain_instance(new_data_scaled[0], best_classifier.predict_proba, num_features=10)
exp.show_in_notebook(show_all=False)
```
